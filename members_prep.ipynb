```python
members_linear = pd.get_dummies(members,columns=['city','gender','registered_via'])
```

```python
members_linear = members_linear.drop('expiration_date',axis=1)
```

```python
members_linear.bd.value_counts()
```

```python
birth_notreal = members_linear.bd[members_linear.bd < 0]
birth_notreal = birth_notreal.value_counts().sum()
birth_notreal
```

```python
birth_notreal2 = members_linear.bd[members_linear.bd > 80]
birth_notreal2 = birth_notreal2.value_counts().sum()
birth_notreal2
```

```python
birth_notreal0 = members_linear.bd[members_linear.bd == 0]
birth_notreal0 = birth_notreal0.value_counts().sum()
birth_notreal0
```

```python
bd_mean = members_linear.bd.sum()/(3000 - members_linear.bd.isnull().sum() - birth_notreal - birth_notreal2 - birth_notreal0)
```

```python
bd_mean
```

```python
members_linear.loc[members_linear.bd == 0, 'bd'] = bd_mean
```

```python
members_linear.to_csv("members_linear.csv", index=False)
```
