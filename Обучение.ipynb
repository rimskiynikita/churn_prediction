```python
import pandas as pnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression 
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
```

```python
def get_acc_1(model, x_train, y_train, x_test, y_test):
    
    model.fit(x_train, y_train)
    pr_y=model.predict(x_test)
    acc_test=accuracy_score(pr_y,y_test)
    acc_train=accuracy_score(model.predict(x_train),y_train)
    
    return  [acc_train, acc_test]
    
```

```python
def get_acc_2(model, x_train, x_test, y_train, y_test):
    
    #model.fit(x_train, y_train)
    pr_y=model.predict(x_test)
    acc_test=accuracy_score(pr_y,y_test)
    acc_train=accuracy_score(model.predict(x_train),y_train)
    
    return [acc_train, acc_test]
    
```

```python
from sklearn.model_selection import GridSearchCV 
from sklearn.model_selection import StratifiedKFold
```

```python
rs=123
```

# Логичтическая

```python
d=pnd.read_csv('linear_model6000.csv')
y=d.is_churn.values
d
```

```python
d_x=d.drop('is_churn', axis=1)
x=d_x.values[:,1:-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
```

```python
penalty_array=['l1', 'l2']
c_array=np.logspace(-5,5, 10)
solver_array=[ 'liblinear'] #'newton-cg' 'lbfgs', 'sag',
max_iter_array=[10,50,100,200]
```

```python
lr_params={'penalty': penalty_array, 'C': c_array, 'solver': solver_array, 'max_iter': max_iter_array}

lr=LogisticRegression(random_state=rs)
skf = StratifiedKFold(n_splits=4, random_state=rs)

lr_grid=GridSearchCV(lr, lr_params, scoring='accuracy', cv=skf)
lr_grid.fit(x_train, y_train)
```

```python
get_acc_2(model=lr_grid,x_train= x_train, x_test=x_test, y_train= y_train, y_test= y_test)
```
