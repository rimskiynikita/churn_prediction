```python
import numpy as np
import pandas as pnd
```

```python
d=pnd.read_csv("user_logs_all.csv")
```

```python
#добавили столбец сумма всех включенных песен
d['total']=d.num_100+d.num_985+d.num_75+d.num_50+d.num_25
d
```

```python
years=[]
months=[]
days=[]

list_of_dates=d.date.tolist()
for i in range(len(list_of_dates)):
    years.append(str(list_of_dates[i])[0:4])
    months.append(str(list_of_dates[i])[4:6])
    days.append(str(list_of_dates[i])[6:8])

d['years']=years
d['months']=months
d['days']=days

d
```

```python
#взяли уникальные имена всех пользователей из таблицы

user_names=[]
b=user_names
for i in d.msno.tolist():
    if i not in user_names:
        user_names.append(i)
```

```python
#посчитал сумму для num25

sums_num_25=[]
for i in range(len(b)):
    a=d[d.msno==user_names[i]]
    sums_num_25.append(sum(a.num_25.tolist()))
```

```python
#функция для подсчета сумм

def sum_nums(column_name):
    sums_num=[]
    for i in range(len(b)):
        a=d[d.msno==b[i]]
        sums_num.append(sum(a[column_name].tolist()))
    return sums_num
```

```python
#функция для подсчета средних

def avg_nums(column_name):
    avgs_num=[]
    for i in range(len(b)):
        a=d[d.msno==b[i]]
        avgs_num.append(sum(a[column_name].tolist())/len(a))
    return avgs_num
```

```python
def num_of_logs():
    num_logs=[]
    for i in range(len(b)):
        a=d[d.msno==b[i]]
        num_logs.append(len(a))
    return num_logs
```

```python
def percentages_of(df,column1, column2):
    
    column1=np.array(df[column1].tolist())
    column2=np.array(df[column2].tolist())
    
    perc=column1/column2
    
    return perc
```

```python
def get_avg_logs_per_month(num_of_month, num_og_logs):
    return num_of_month/num_of_logs
```

```python
new_d= pnd.DataFrame()
new_d['user_name']=user_names
new_d['num_of_logs']=num_of_logs()
new_d['sum_num25']=sum_nums(column_name="num_25")
new_d['avg_num25']=avg_nums(column_name='num_25')
new_d['sum_num50']=sum_nums(column_name="num_50")
new_d['avg_num50']=avg_nums(column_name='num_50')
new_d['sum_num75']=sum_nums(column_name="num_75")
new_d['avg_num75']=avg_nums(column_name='num_75')
new_d['sum_num985']=sum_nums(column_name="num_985")
new_d['avg_num985']=avg_nums(column_name='num_985')
new_d['sum_num100']=sum_nums(column_name="num_100")
new_d['avg_num100']=avg_nums(column_name='num_100')
new_d['sum_num_total']=sum_nums(column_name="total")
new_d['avg_num_total']=avg_nums(column_name='total')
new_d['sum_num_total']=sum_nums(column_name="total")
new_d['avg_num_total']=avg_nums(column_name='total')
new_d['sum_num_unq']=sum_nums(column_name="num_unq")
new_d['avg_num_unq']=avg_nums(column_name='num_unq')
new_d['sum_total_secs']=sum_nums(column_name="total_secs")
new_d['avg_total_secs']=avg_nums(column_name='total_secs')
new_d['perc_of_accidentally_songs']=percentages_of(new_d, 'sum_num25', 'sum_num_total')
new_d['perc_of_favourite_songs']=percentages_of(new_d, 'sum_num100', 'sum_num_total')

new_d
```
