### LINEAR MODELS

```python
members_lin = pd.read_csv('members_linear6000.csv')
trans_lin = pd.read_csv('transactions_linear6000.csv')
logs_lin = pd.read_csv('logs_linear6000.csv')
```

```python
trans_lin.head()
logs_lin.columns
```

```python
# import matplotlib.pyplot as plt

# for i in logs_lin.drop('msno', axis=1).columns:
#     print(i)
#     logs_lin[i].hist()
#     plt.show()
```

```python
# logs_lin['num_of_logs'] = logs_lin.num_of_transactions
# logs_lin.drop('num_of_transactions', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler1 = StandardScaler()#MinMaxScaler(feature_range=(0, 1))
members_lin.bd = scaler1.fit_transform(members_lin.bd.reshape(-1,1))
logs_log = ['sum_num25', 'avg_num25', 'sum_num50', 'avg_num50',
             'sum_num75', 'avg_num75', 'sum_num985', 'avg_num985', 'sum_num100',
             'avg_num100', 'sum_num_unq', 'avg_num_unq', 'sum_total_secs',
             'avg_total_secs', 'sum_total_songs', 'avg_total_songs', 'avg_songs_per_month', 
             'min_days_diff', 'num_of_logs', 'perc_of_accidentally_songs', 'perc_of_favourite_songs', 
             'avg_days_diff', 'max_days_diff']

for i in logs_log:
    logs_lin[i] = scaler1.fit_transform(logs_lin[i].reshape(-1,1))
trans_log = ['num_of_transactions', 'extra_payment_plans_total', 'extra_payment_plans_count', 
        'avg_plan_list_price', 'avg_actual_amount_paid', 'is_cancel_total', 'left_payment_days_total', 
        'avg_transaction_days_diff', 'max_transaction_days_diff', 'min_transaction_days_diff', 
        'avg_expire_days_diff', 'max_expire_days_diff', 'min_expire_days_diff']

for i in trans_log:
    trans_lin[i] = scaler1.fit_transform(trans_lin[i].reshape(-1,1))
```

```python
logs_lin.head()
```

```python
unique_ids = members_lin.msno.unique()
```

```python
train = train[train.msno.isin(unique_ids)]
```

```python
lin1 = pd.merge(members_lin, logs_lin, on='msno')
```

```python
lin2 = pd.merge(lin1, trans_lin, on='msno')
```

```python
lin_final = pd.merge(lin2, train, on='msno')
```

```python
lin_final.to_csv('linear_model6000.csv', index=False)
```

```python
print(lin_final.shape)
lin_final.head()
```

```python
for i in lin_final.columns:
    print(i)
```

```python
scaler2 = MinMaxScaler(feature_range=(0, 1))
# members_xg.bd = scaler2.fit_transform(members_xg.bd.reshape(-1,1))
members_tree.bd = scaler2.fit_transform(members_tree.bd.reshape(-1,1))
trans_log = ['num_of_transactions', 'extra_payment_plans_total', 'extra_payment_plans_count', 
        'avg_plan_list_price', 'avg_actual_amount_paid', 
        'is_cancel_total', 'left_payment_days_total', 
        'avg_transaction_days_diff', 'max_transaction_days_diff', 'min_transaction_days_diff', 
        'avg_expire_days_diff', 'max_expire_days_diff', 'min_expire_days_diff']

for i in logs_log:
    logs_xg[i] = scaler2.fit_transform(logs_xg[i].reshape(-1,1))
    
logs_log = ['sum_num25', 'avg_num25', 'sum_num50', 'avg_num50',
             'sum_num75', 'avg_num75', 'sum_num985', 'avg_num985', 'sum_num100',
             'avg_num100', 'sum_num_unq', 'avg_num_unq', 'sum_total_secs',
             'avg_total_secs', 'sum_total_songs', 'avg_total_songs', 'avg_songs_per_month', 
             'min_days_diff', 'num_of_logs', 'perc_of_accidentally_songs', 'perc_of_favourite_songs', 
             'avg_days_diff', 'max_days_diff']
for i in trans_log:
    trans_xg[i] = scaler2.fit_transform(trans_xg[i].reshape(-1,1))
```
